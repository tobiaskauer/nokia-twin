```python
import pandas as pd
import sqlalchemy
import tqdm
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String
from IPython.display import display

engine = sqlalchemy.create_engine('mysql+mysqlconnector://root@localhost/nokiatwin')
```

```python
haystack = pd.read_csv('/Users/tk/Documents/GitHub/nokia/nokia-twin/assets/country_regions_new.tsv',sep='\t')
haystack
```

```python
reviews = pd.read_sql_query("SELECT DISTINCT location, country FROM reviews_realign where country is NULL",
#reviews = pd.read_sql_query("SELECT DISTINCT location, country FROM reviews",
    engine)

for index, row in reviews.iterrows():
    needle = row['location']
    result = haystack.loc[haystack['location'] == needle]
    #display(result.values)
    if len(result.values) > 0:
        reviews.at[index,"country"] = result['country'].values.max()
        #reviews.at[index,"region"] = result['region'].values.max()
reviews.head(20)
```

```python
#update reviews that have a new entry in the DB

meta = MetaData()
students = Table(
   'reviews_realign', 
   meta, 
   Column('location', String), 
   Column('country', String), 
   #Column('region', String), 
)

conn = engine.connect()

for index, row in tqdm(reviews.iterrows(), total=reviews.shape[0]):
#for index, row in reviews.iterrows():
    if row["country"] is not None:
        
        #updated = {"country": row['country'], "region": row['region']}
        updated = {"country": row['country']}

        stmt=students.update().where(students.c.location==row['location']).values(updated)
        conn.execute(stmt)
```

```python
#show what locations still have no country, ordered by how often they appear
df = pd.read_sql_query("SELECT location,COUNT(*) as count FROM reviews       WHERE country is null GROUP BY location ORDER BY count desc",
    engine)

for index, row in df.iterrows():
    print(row["location"])
```
